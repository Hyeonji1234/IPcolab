```python
img = cv2.imread('/content/gdrive/My Drive/digital/Images/gsy.jpg')
print(img.shape)
gimg1 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
#plt.imshow(gimg1, cmap='gray')
cv2_imshow(gimg1)
```

```python
outimage = gimg1.copy()
#bluriamge = cv2.GaussianBlur(outimage, (7,7), 5)
bluriamge = cv2.medianBlur(outimage, 7)

x = 265
y = 260
size = 15

for h in range(y, y+size):
  for w in range(x, x+size):
    #outimage[h][w] = 255
    outimage[h][w] = bluriamge[h][w]

cv2_imshow(outimage)
```

```python
img = cv2.imread('/content/gdrive/My Drive/digital/Images/lena256.jpg')
print(img.shape)
gimg1 = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
#plt.imshow(gimg1, cmap='gray')
cv2_imshow(gimg1)
```

```python
outimg = cv2.GaussianBlur(gimg1, (7,7), 2)
cv2_imshow(outimg)
```

```python
edgeimage = cv2.addWeighted(gimg1, 1, outimg, -1, 0)
cv2_imshow(edgeimage)
```

```python
unsharpimage = cv2.addWeighted(gimg1, 1, outimg, 1, 1)
cv2_imshow(unsharpimage)
```

```python
kernel3 = np.ones((3,3), np.float32)/9
kernel3[0][0] = -1/9
kernel3[0][1] = -1/9
kernel3[0][2] = -1/9
kernel3[1][0] = -1/9
kernel3[1][1] =  17/9
kernel3[1][2] = -1/9
kernel3[2][0] = -1/9
kernel3[2][1] = -1/9
kernel3[2][2] = -1/9
```

```python
output = cv2.filter2D(gimg1, -1, kernel3)
cv2_imshow(output)
```
